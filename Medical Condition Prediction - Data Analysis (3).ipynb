
# <div style='background:darkblue'>  <font color='white'> <b>&nbsp;Exploratory Data Analysis</b></font> </div>

```python
# gender percentage (pie chart)
df['gender'].value_counts().plot.pie(autopct='%.2f%%', title='Gender')
plt.show()
```

```python
#total gender medical condition (countplot)
a = sns.countplot(x='gender',data=df,palette='bright',hue='gender')
for i in a.containers:
    a.bar_label(i,)
plt.title('Total Males and Females medical conditions')
plt.xlabel('Gender')
plt.ylabel('Total number of persons')
plt.show()
```

```python
#total smoking status (countplot)
a=sns.countplot(x='smoking_status',data=df,palette='bright',hue='smoking_status')
for i in a.containers:
    a.bar_label(i,)
plt.title('Total smoking_status')
plt.ylabel('Total number of persons')
plt.show()
```

```python
#types of diseases (countplot)
a=sns.countplot(x='condition',data=df,palette='bright',hue='condition')
for i in a.containers:
    a.bar_label(i,)
plt.title('Types Of Diseases')
plt.xlabel('medical conditions')
plt.ylabel('Total number of persons')

plt.show()
```

```python
#average blood pressure by age (lineplot)
ageblood = df.groupby('age')['blood_pressure'].mean().reset_index()
plt.figure(figsize=(14,6))
sns.lineplot(x="age", y="blood_pressure", data=ageblood, color="blue", linewidth=2.5, marker='o')

# Annotating each point with the value of blood pressure
for x, y in zip(ageblood['age'], ageblood['blood_pressure']):
    plt.text(x, y, f'{y:.0f}', ha='center', va='bottom', fontsize=10, color='black')

plt.title('Average Blood Pressure by Age')
plt.xlabel('Age')
plt.ylabel('Blood Pressure')
plt.grid()
plt.show()
```

 Not much trend between blood pressure and age.

```python
# average glucose levels by age (lineplot)
agegluc = df.groupby('age')['glucose_levels'].mean().reset_index()

# Create the figure and plot the lineplot
plt.figure(figsize=(14,6))
sns.lineplot(x="age", y="glucose_levels", data=agegluc, color="blue", linewidth=2.5, marker='o')

# Annotate each point with its value
for x, y in zip(agegluc['age'], agegluc['glucose_levels']):
    plt.text(x, y, f'{y:.0f}', ha='center', va='bottom', fontsize=10, color='black')

plt.title('Average Glucose Levels by Age')
plt.xlabel('Age')
plt.ylabel('Glucose Levels')
plt.grid()
plt.show()
```

* Not much trend between age and glucose levels

```python
# average body mass index by age (lineplot)
agebmi = df.groupby('age')['bmi'].mean().reset_index()
# Create the figure and plot the lineplot
plt.figure(figsize=(14,6))
sns.lineplot(x="age", y="bmi", data=agebmi, color="blue", linewidth=2.5, marker='o')
# Annotate each point with its BMI value
for x, y in zip(agebmi['age'], agebmi['bmi']):
    plt.text(x, y, f'{y:.1f}', ha='center', va='bottom', fontsize=10, color='black')
plt.title('Average Body Mass Index by Age')
plt.xlabel('Age')
plt.ylabel('BMI')
plt.grid()
plt.show()
```

* Not much trend between age and BMI either.

```python
# Average Blood Pressure by Smoking Status (barplot)
smokeblood = df.groupby('smoking_status')['blood_pressure'].mean().reset_index()
plt.figure(figsize=(8,5))
a=sns.barplot(x='smoking_status', y='blood_pressure', data=smokeblood,palette='bright',hue='smoking_status')
for i in a.containers:
    a.bar_label(i)
plt.ylabel('Blood Pressure')
plt.xlabel('Smoking Status')
plt.title('Average Blood Pressure by Smoking Status')
plt.show()
```

* Average blood pressure seems to be even for both non-smokers and smokers.

```python
# Average Glucose Levels by Smoking Status (barplot)
smokeblood = df.groupby('smoking_status')['glucose_levels'].mean().reset_index()
plt.figure(figsize=(8,5))
a=sns.barplot(x='smoking_status', y='glucose_levels', data=smokeblood,palette='bright',hue='smoking_status')
for i in a.containers:
    a.bar_label(i)
plt.title('Average Glucose Levels by Smoking Status')
plt.ylabel('Glucose Level')
plt.xlabel('Smoking Status')
plt.show()
```

* Average glucose levels also seem to be the same for both statuses.

```python
# Diseases of Smoking Status (countplot)
plt.figure(figsize=(10,6))
plt.title("Diseases of Smoking Status")
plt.xlabel('Medical Condition')
a=sns.countplot(x='condition', hue='smoking_status', data=df)
for i in a.containers:
    a.bar_label(i)
plt.show()
```

* More non-smokers in all three conditions, especially in diabetes.

```python
# Distribution of Smoking Status by Medical Condition (countplot)
plt.figure(figsize=(10,6))
plt.title("Distribution of Smoking Status by Medical Condition")
plt.xlabel('Smoking Status')
a=sns.countplot(x='smoking_status', hue='condition', data=df)
for i in a.containers:
    a.bar_label(i)
plt.show()
```

* Majority of non-smokers surveyed seem to have diabetes while diabetes is also popular among the small amount of smokers surveyed.

```python
# Distribution of Condition by Gender (countplot)
plt.figure(figsize=(10,6))
plt.title("Distribution of Condition by Gender")
a=sns.countplot(x='condition', hue='gender', data=df)
for i in a.containers:
    a.bar_label(i)
plt.show()
```

```python
# Gender Distribution by Condition (countplot)
a=sns.countplot(x='gender', hue='condition', data=df)
plt.title('Gender Distribution by Condition')
for i in a.containers:
    a.bar_label(i)
plt.show()
```

```python
# Distribution of Smoking Status by Gender (countplot)
plt.figure(figsize=(10,6))
plt.title("Distribution of Smoking Status by Gender")
a=sns.countplot(x='smoking_status', hue='gender', data=df)
for i in a.containers:
    a.bar_label(i)
plt.show()
```

* This data might be biased as the amount of females in this data set is extremely small compared to males
